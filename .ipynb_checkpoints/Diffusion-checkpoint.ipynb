# Diffusion

This file implements the diffusion equation in a 2d system. It creates an animation of the concentration at points in the system over a set time interval. For times under a certain cutoff the one-dimensional profile along the center of the system is fitted to a Gaussian curve and the standard deviation of these curves is plotted together with its theory function against time. Furthermore the sum of the system (a measure of how many particles there are in the system) is plotted against time.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import matplotlib
from scipy.stats import norm
from scipy.optimize import curve_fit
matplotlib.rcParams['animation.embed_limit'] = 100000000
```

```python
# Defining functions to be used later on


def gaussian(x, a, b, c):
    """Gaussian function used for fitting data. To be used together
    with scipy.optimize.curve_fit.

    Keyword Arguments:
    -x :  Numpy array of positions used for computing.
    -a :  Height of peak.
    -b :  Position of center of the peak.
    -c :  Standard deviation.
    """
    return a*np.exp(-(x-b)**2/(2*c**2))


def theory(t):
    """Theory function describing the change in the standard deviation
    of the concentration ditribution over time.

    Keyword Arguments:
    -t :  Numpy array of time values used for computing.

    Required variables:
    -D :  Rate of diffusion.
    """
    global D
    return np.sqrt(2 * D * t)
```

```python
# Setting up inital parameters
t0 = 0  # Starting time (s)
tf = 100  # End time (s)
fitcutoff = 10  # Time until which data is collected for fitting (s)
dt = 0.01  # Timestep used. Important for diffusion model (s)
D = 0.001  # Rate of diffusion.
Nparticles = 10  # Number of particles in system
N = 101  # Number of points sampled along each axis
# I am using 101 points so that the initial particles are
# perfectly centered
tperframe = 0.1  # Time passed before a new frame is created

# Defining dimensions of system, in this cas a 1x1 square
xmin, xmax, ymin, ymax = 0, 1, 0, 1
# Computing the change in x and y when moving across the system
dx, dy = xmax/N, ymax/N

# Setting up the system
C = np.zeros((N, N))

# Creating a copy of the system to write new system data to
newC = np.copy(C)

# Adding the particles in the center
C[N//2, N//2] = Nparticles

# Creating figure for animation
fig = plt.figure(figsize=(8, 8))

# Creating lists for:
# Saving frames for animation (ims), including initial
# Saving the center row of system for fitting (fitdata)
# Saving the sum of the system after each cycle (sums)
ims, fitdata, sums = [(plt.pcolormesh(np.copy(C), vmin=0), )], [], []

timedata = []

for f in range(int(tf / dt)):
    # Computing second derivative of diffusion equation
    d2Cdx2 = (C[2:, 1:-1] + C[:-2, 1:-1] - 2 * C[1:-1, 1:-1]) / (dx**2)
    d2Cdy2 = (C[1:-1, 2:] + C[1:-1, :-2] - 2 * C[1:-1, 1:-1]) / (dy**2)
    # Computing the new system state after one cycle
    newC[1:-1, 1:-1] = C[1:-1, 1:-1] + D * dt * (d2Cdx2 + d2Cdy2)
    # Taking care of boundary condiditons (reflective)
    newC[0, :] = newC[1, :]
    newC[-1, :] = newC[-2, :]
    newC[:, 0] = newC[:, 1]
    newC[:, -1] = newC[:, -2]
    C, newC = newC, C
    if f <= fitcutoff / dt:
        fitdata.append(np.copy(C[:, N//2]))
        timedata.append(theory(f * dt))
    if f % tf / tperframe == 0:
        ims.append((plt.pcolormesh(np.copy(C), vmin=0), ))
    sums.append(np.abs(np.sum(np.copy(C))))

# Creating the animation and finish setting up figure
ani = animation.ArtistAnimation(fig, ims, interval=20, repeat=False)

plt.title(f'Diffusion Equation applied to 2d system from {t0}s to {tf}s')
plt.tight_layout()
plt.close()

# Displaying the animation
HTML(ani.to_jshtml())
```

```python
# Defining guesses for curve_fit
guesses = (1, xmax//2, 1)

# Creating x-space for the sampled data
xsample = np.linspace(0, 1, N)

stdevs = []  # List for saving the standard deviations of the fitted curves to

# Loop for extracting the fitdata and performing the fit
for sample in fitdata:
    # Fitting gaussian() to the data
    fitparams, covariance = curve_fit(gaussian, xsample, sample, p0=guesses)
    # Saving the absolute value of standard deviation in order to account
    # for negative values coming from the fit parameters
    stdevs.append(np.abs(fitparams[2]))

# Creating timespace for plotting the sample data and theory function
datatime = theory(np.linspace(t0, fitcutoff, len(stdevs)))
ttheory = np.linspace(t0, theory(fitcutoff), 1000)

# Scaling plot/evenly spacing points
evenstdev, eventimes = [], []

for i in range(len(stdevs)):
    if i < np.sqrt(len(stdevs)):
        evenstdev.append(stdevs[i**2])
        eventimes.append(datatime[i**2])

# Setting up the plot
plt.figure(figsize=(13, 6.5))
plt.title('stdev vs sqrt(2*D*Time)')
plt.xlabel('sqrt(2*D*Time)')
plt.ylabel('stdev')
# Omitting the first couple points, as they tend towards infinity
plt.plot(eventimes[2:], evenstdev[2:], 'g.', label='data')
plt.plot(ttheory, ttheory, label='Theory')
plt.legend()
plt.show()
```

```python
# Creating timespace to plot
time = np.linspace(t0, tf, len(sums))

plt.figure(figsize=(13, 6.5))
plt.title('Sum of concentration across dish over time')
plt.plot(time, sums)
plt.xlabel('Time (s)')
plt.ylabel('Sum of concentration accross array')
plt.show()
```
